```python
mean_age_survived = data.groupby("Survived")["Age"].mean()
mean_age_survived
```

```python
def fill_age(row):
    Age = row["Age"]
    Survived = row["Survived"]
    # print(f"Processing row: Age={Age}, Survived={Survived}")
    
    if pd.isna(Age):  
        if Survived == 0:
            # print("Replacing with mean for Survived=0")
            return mean_age_survived[0]
        elif Survived == 1:
            # print("Replacing with mean for Survived=1")
            return mean_age_survived[1]
    else:
        # print("Age not NaN, keeping original value")
        return Age
```

```python
# Apply the function to fill NaN values
data["Age"] = data.apply(fill_age, axis=1)
```

### Split data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, ConfusionMatrixDisplay
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(data.drop("Survived" , axis=1 ), data["Survived"] , test_size=0.30 , random_state = 101)
```

## Training and predicting 

```python
logModel = LogisticRegression(solver='liblinear', max_iter=100)
logModel.fit(X_train , Y_train)
```

```python
predictions = logModel.predict(X_test)
```

```python
conf_matrix = confusion_matrix(Y_test , predictions)
```

```python
conf_matrix
```

```python
# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=logModel.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
# Calculate performance metrics
accuracy = accuracy_score(Y_test, predictions)
precision = precision_score(Y_test, predictions, average='binary')  # Adjust 'binary' for multi-class cases
recall = recall_score(Y_test, predictions, average='binary')
```

```python
# Print metrics
print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
```
