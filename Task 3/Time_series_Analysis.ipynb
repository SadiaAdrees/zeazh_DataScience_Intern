```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
```

```python
print('Dickey Fuller Test Results:')
test_data= adfuller(resample_data.iloc[:,0].values, autolag='AIC')
data_output = pd.Series(test_data[0:4], index=['Text Statistic', 'p-value', 'Lags Used', 'Number of observations'])
for key, value in test_data[4].items():
    data_output['Critical Value (%s)'%key] = value
print(data_output)
```

```python
decomp = seasonal_decompose(resample_data, period=3)

trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid
```

```python
fig = make_subplots(rows=4, cols=1, subplot_titles=('Original', 'Trend', 'Seasonal', 'Residual'))


fig.add_trace(go.Scatter(x=resample_data.index, y=resample_data['AvgTemp'], mode='lines', name='Original'), row=1, col=1)


fig.add_trace(go.Scatter(x=trend.index, y=trend, mode='lines', name='Trend'), row=2, col=1)


fig.add_trace(go.Scatter(x=seasonal.index, y=seasonal, mode='lines', name='Seasonal'), row=3, col=1)


fig.add_trace(go.Scatter(x=residual.index, y=residual, mode='lines', name='Residual'), row=4, col=1)

fig.update_layout(title='Temperature Decomposition',
                  xaxis_title='Year',
                  yaxis_title='Temperature')
fig.show()
```

```python
rol_mean = resample_data.rolling(window=3, center=True).mean()
ewm = resample_data.ewm(span=3).mean()
rol_std = resample_data.rolling(window=3, center=True).std()

print('Rolling mean', rol_mean.head())
print('Exponentially weighted mean', ewm.head())
print('Rolling standard deviation', rol_std.head())
```

```python
# Creating subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=('Original vs. Rolling Mean vs. Exponentially Weighted Mean', 'Rolling Standard Deviation'))

# Plotting Original vs. Rolling Mean vs. Exponentially Weighted Mean
fig.add_trace(go.Scatter(x=resample_data.index, y=resample_data['AvgTemp'], mode='lines', name='Original'), row=1, col=1)
fig.add_trace(go.Scatter(x=rol_mean.index, y=rol_mean['AvgTemp'], mode='lines', name='Rolling Mean'), row=1, col=1)
fig.add_trace(go.Scatter(x=ewm.index, y=ewm['AvgTemp'], mode='lines', name='Exponentially Weighted Mean'), row=1, col=1)

# Plotting Rolling Standard Deviation
fig.add_trace(go.Scatter(x=rol_std.index, y=rol_std['AvgTemp'], mode='lines', name='Rolling STD'), row=1, col=2)

# Updating layout
fig.update_layout(title='Temperature Changes from 1980-2013', xaxis_title='Year', yaxis_title='Temperature')
fig.update_xaxes(title_text='Year', row=1, col=1)
fig.update_yaxes(title_text='Temperature', row=1, col=1)
fig.update_xaxes(title_text='Year', row=1, col=2)
fig.update_yaxes(title_text='Temperature', row=1, col=2)

fig.show()
```

```python
train_size = int(len(resample_data) * 0.8)
train, test = resample_data.iloc[:train_size], resample_data.iloc[train_size:]
```

```python
model = ARIMA(train, order=(5, 1, 0))
fit_model = model.fit()
```

```python
predictions = fit_model.forecast(steps=len(test))
predictions
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rmse = sqrt(mean_squared_error(test, predictions))
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=train.index, y=train['AvgTemp'], mode='markers,lines', name='Train'))
fig.add_trace(go.Scatter(x=test.index, y=test['AvgTemp'], mode='markers,lines', name='Test'))
fig.add_trace(go.Scatter(x=test.index, y=predictions, mode='markers,lines', name='Predictions'))
fig.update_layout(title='ARIMA Model Forecast',
                  xaxis_title='Date',
                  yaxis_title='Temperature',
                  legend=dict(x=0, y=1, traceorder='normal'))
fig.show()
```

```python
train_size = int(len(resample_data) * 0.8)
train, test = resample_data.iloc[:train_size], resample_data.iloc[train_size:]
```

```python
best_rmse = float('inf')
best_order = None

for p in range(1, 5):
    for d in range(1, 3):
        for q in range(1, 5):
            order = (p, d, q)
            model = ARIMA(train, enforce_invertibility=False, order=order)
            fit_model = model.fit()
            predictions = fit_model.forecast(steps=len(test))
            rmse = sqrt(mean_squared_error(test, predictions))

            if rmse < best_rmse:
                best_rmse = rmse
                best_order = order
```

```python
final_model = ARIMA(resample_data, order=best_order)
fit_model = final_model.fit()
```

```python
predictions = fit_model.forecast(steps=len(test))
```

```python
rmse = sqrt(mean_squared_error(test, predictions))
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=train.index, y=train['AvgTemp'], mode='markers,lines', name='Train'))
fig.add_trace(go.Scatter(x=test.index, y=test['AvgTemp'], mode='markers,lines', name='Test'))
fig.add_trace(go.Scatter(x=test.index, y=predictions, mode='markers,lines', name='Predictions'))
fig.update_layout(title='ARIMA Model Forecast',
                  xaxis_title='Date',
                  yaxis_title='Temperature',
                  legend=dict(x=0, y=1, traceorder='normal'))


fig.show()
```
